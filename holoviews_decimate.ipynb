**IMPORT RESULTS OF PERIODICITY DETECTION**

Imported file cotnains:

*  *name* is quasar ID in AGN Data Challenge data base
*   *m1* and *m2* are mean sampling rate in given bands where period is detected


*   *m3* is detected period in given pair of bands, when period is detected in two bands we requiere that detected values in two bands differes less then 10% of relative error
*   *m4* and *m5* are lower and upper error of  detected period
*   *m6* is significance of detected period. Significance is determined via Johnson shuffling method
*   *m7* is pair of bands where period is detected: bands are designed as **G=1, BP=2,RP=3,RP=4**, and pairs as **GBP=12, GRP=13, GRP=14**. We chose to compare results with respect to **G*
* band as we expect it to be less deformed of all bands.

```python
frames = [data1,data2,data2,data3,data4,data5]

data_comb = pd.concat(frames)
```

```python
# Define a threshold
threshold = 0.25

# Find the rows where the absolute difference between detected period ('m3') and mean samplings('m1' or 'm2') is less than 25% of 'm3'
mask = (abs(data_comb['m3'] - data_comb['m1']) / data_comb['m3'] < threshold) | (abs(data_comb['m3'] - data_comb['m2']) / data_comb['m3'] < threshold)

# Discard those rows
filtered_df = data_comb[~mask]
```
